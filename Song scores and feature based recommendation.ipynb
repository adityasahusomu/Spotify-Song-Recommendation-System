```python
num_cols = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness','valence', 'tempo']
```

```python
def ohe_preprocessing(df, column, new_name):
    ohe_df = pd.get_dummies(df[column])
    features = ohe_df.columns
    ohe_df.columns = [new_name + '|' + str(i) for i in features]
    ohe_df.reset_index(drop=True, inplace=True)
    return ohe_df
```

```python
# Tf-Idf matrix

from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer
def create_features_df(df, col_list):
    tqdm.pandas()
    
    tfidf = TfidfVectorizer()
    print("Processing TF-IDF...")
    tfidf_matrix = tfidf.fit_transform(df['genres_upd'].progress_apply(lambda x: ' '.join(x))).toarray()
    df_genre_concat1 = pd.DataFrame(tfidf_matrix)
    df_genre_concat1.columns = ['genre' + '|' + str(i) for i in tfidf.get_feature_names_out()]

    
    print("Processing one-hot encoding for 'year'...")
    year_ohe = ohe_preprocessing(df, 'year', 'year')* 0.5
    
    print("Processing one-hot encoding for 'popularity_bin'...")
    popularity_ohe = ohe_preprocessing(df, 'popularity_bin', 'popl') * 0.15
    
    print("Scaling numerical columns...")
    scaler = MinMaxScaler()
    num_cols_scaled = pd.DataFrame(scaler.fit_transform(df[col_list].reset_index(drop=True)), columns=col_list) * 0.2
    
    # Concatenate all features...
    print("Concatenating all features...")
#     final = pd.concat([df_genre_concat1, year_ohe, popularity_ohe, num_cols_scaled], axis = 1, copy=False)
    final_chunks = []
    for start in range(0, len(df), 1000):
        end = start + 1000
        chunk = pd.concat([df_genre_concat1.iloc[start:end], 
                           year_ohe.iloc[start:end], 
                           popularity_ohe.iloc[start:end], 
                           num_cols_scaled.iloc[start:end]], axis=1, copy=False)
        final_chunks.append(chunk)
    
    final = pd.concat(final_chunks, axis=0, copy=False)
    
    
    print('operation ends..')

    return final
```

```python
complete_feature_set = create_features_df(spotify_df_exploded_merged_genre, col_list = num_cols)
complete_feature_set.head()
```

```python
matrix = complete_feature_set.values

# Creating playlist Vectors
from sklearn.metrics.pairwise import cosine_similarity
similar = cosine_similarity(matrix)
```

```python
similar.shape
```

```python
spotify_df_exploded_merged_genre.head(2)
```

```python
# Recommendation function

def recommender(song_name):
    idx = spotify_df_exploded_merged_genre[spotify_df_exploded_merged_genre['name_x']==song_name].index[0]
    distance = sorted(list(enumerate(similar[idx])),reverse=True, key=lambda x: x[1])    # distance is a list of tuples. Each tuple contains index and the similiarity score
    
    song = []
    for s_id in distance[1:15]:
        song.append(spotify_df_exploded_merged_genre.iloc[s_id[0]].name_x)
    return song
```
